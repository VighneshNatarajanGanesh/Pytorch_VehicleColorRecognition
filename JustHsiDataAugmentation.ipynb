```python
import numpy as np
import h5py
import os
```

make a new folder with individual clannels as images with the prefix 'xx_' where xx represnts channel number

# 10 channels

test

```python
input_hsi_dir = os.path.join('datasets', 'HsiData', 'test')
image_list = os.listdir(input_hsi_dir)
path = sorted([os.path.join(input_hsi_dir, f) for f in image_list if f.endswith('hdf5')])
print("no of images:", len(path))
```

original channel : new channel no

02:1 

05:2

08:3

11:4

14:5 

17:6

20:7

23:8

26:9

29:10

```python
HDF5_DATASET_NAME = 'hs_data'
INPUT_CHANNELS = [2,5,8,11,14,17,20,23,26,29]

print("Processing:", end=" ")

for file in path:
    with h5py.File(file, 'r') as f:
        data = f[HDF5_DATASET_NAME][:].astype(np.float32)
        

    for i in range(10):
        t_path = file.split("/")
        t_path[-1] = "{:0>2d}".format(i)+"_"+t_path[-1]
        t_path[-3] = t_path[-3]+"10"
        t_path = "/".join(t_path)
        hf = h5py.File(t_path, 'w')
        hf.create_dataset('hs_data', data=data[:,:,INPUT_CHANNELS[i]])
        hf.close()
    print(".", end="")
```
